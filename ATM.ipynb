```python
class ATM:
    def __init__(self,initial_balance=0):
        self.account_no = int(input('Enter your account number :'))
        self.name = input('Enter your name :')
        self.balance = initial_balance
        self.limit = 20000
        
    def Balance_check(self):
        print(f'total amount in your account having account no. {self.account_no} is {self.balance}\n')
    
    def Deposit(self):
        amount = int(input('Enter the amount : '))
        self.balance += amount
        print(f'{amount} is added to your account having account no. {self.account_no}\n')
    
    def Withdraw(self):
        amount = int(input('Enter the amount'))
        if(self.balance < amount):
            print('you have not sufficient balance \n')
        elif self.limit < amount:
            print('Sorry for inconvinient, your amount is excess than limit')
        else:
            self.balance -= amount
            print(f'{amount} is successfully withdraw from you account {self.account_no}\n')
        
    def display(self):
        while True:
            print('Select 1 option which you want to do')
            print('Enter 1 for -------Balance_check')
            print('Enter 2 for -------Deposit money')
            print('Enter 3 for -------Withdraw money')
            print('Enter 4 for -------Exit')

            choice = int(input())
            if choice == 1:
                ATM.Balance_check(self)
            elif choice == 2:
                ATM.Deposit(self)
            elif choice == 3:
                ATM.Withdraw(self)
            else:
                break
        
            

customer1 = ATM()
customer1.display()

        
    
    
            
    
```
